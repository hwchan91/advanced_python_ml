```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm # color map

np.random.seed()

from sklearn.decomposition import PCA
from sklearn.preprocessing import scale

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
data = scale(digits.data)
# data = digits.data
data.shape
labels = digits['target']
```

```python
# k-means clustering
from time import time
from sklearn import metrics
from sklearn.cluster import KMeans

# data = scale(digits.data)

n_samples, n_features = data.shape
n_digits = len(set(labels))
```

```python
def bench_k_means(estimator, name, data, labels, sample_size=300):
    t0 = time()
    estimator.fit(data)
    print('% 9s      %i   %.3f   %.3f   %.3f   %.3f    %.3f'
          % (
                 name, 
                 (time() - t0), 
#                  estimator.inertia_,
                 metrics.homogeneity_score(labels, estimator.labels_),
                 metrics.completeness_score(labels, estimator.labels_),
                 metrics.v_measure_score(labels, estimator.labels_),
                 metrics.adjusted_rand_score(labels, estimator.labels_),
                 metrics.silhouette_score(
                     data, 
                     estimator.labels_,
                     metric='euclidean',
                     sample_size=sample_size
                 )
            )
         )
```

```python
print(79 * '_')  
print(f'n_digits: {n_digits}, \t n_samples {n_samples}, \t n_features {n_features}')  
print(79 * '_')    
print('% 9s' % 'init'
      '         time   homo   compl   v-meas   ARI     silhouette')
bench_k_means(
    KMeans(
        init='k-means++', 
        n_clusters=n_digits, 
        n_init=10
    ), 
    name="k-means++", 
    data=data,
    labels=labels
)
print(79 * '_')
```

```python
# brief explanation of SOM
# http://www.cs.bham.ac.uk/~jxb/NN/l16.pdf
# According to it, X should be normalized first, and cost = (X-w)^2 instead of ||X-w||
```

```python
# from math import sqrt
from numpy import (array, unravel_index, nditer, linalg, random, subtract,
                   power, exp, pi, zeros, arange, outer, meshgrid, dot)
# from collections import defaultdict
from warnings import warn
from numpy.linalg import norm


# def fast_norm(x):
#     """Returns norm-2 of a 1-D numpy array.
#     * faster than linalg.norm in case of 1-D arrays (numpy 1.9.2rc1).
#     """
#     return sqrt(dot(x, x.T))

# note: norm (Frobenius norm) = sqrt(sum of squares of elems in arr), also called euclidean norm

class Som:
    def __init__(self,x,y,input_len,sigma=1.0,learning_rate=0.5,random_seed=None):
       
        if sigma >= x/2.0 or sigma >= y/2.0:
            warn('Warning: sigma is too high for the dimension of the map.')
        
        self.random_generator = random.RandomState(random_seed)
        self.learning_rate = learning_rate
        self.sigma = sigma
        # weights is a 3-dimensional(or dimensions of data + 1) matrix
        # or, easily put, a matrix (x,y) of arrays(of len input_length=n_features)
        # each array is assessible by calling weight[(xi,yi)]
        self.weights = self.random_generator.rand(x,y,input_len)*2-1 # random initialization
        self.weights = array([v/linalg.norm(v) for v in self.weights]) # normalization
        self.activation_map = zeros((x,y)) # var to keep track of the errors of the current sample to each of the weights
        # used to evaluate the neighborhood function
        self.neigx = arange(x) 
        self.neigy = arange(y)


    # by init, weights are randomly initialized
    # calling this will instead set the weights=random samples from data (w repetition)
    def random_weights_init(self,data):
        # note: nditer is an iterator, serving as a loop while saving memory
        # nditer loops through each row, then each col/elem in each row, the elem can be accessed by arr[it.multi_index]
        it = nditer(self.activation_map, flags=['multi_index']) # note: activation map is not actually used in this method, it is only used to get the set of indices within (x,y)
        while not it.finished:
            self.weights[it.multi_index] = self.get_rand_sample(data)
            self.weights[it.multi_index] = self.weights[it.multi_index]/norm(self.weights[it.multi_index]) # normalized
            it.iternext()
            
    def get_rand_sample(self,data):
        rand_i = int(round(self.random_generator.rand()*len(data)-1)) 
        return data[rand_i]
    
    def train_random(self,data,num_iteration):        
        """ Trains the SOM picking samples at random from data """      
        self.T = num_iteration/2 # to adjust learning rate
    
        for iteration in range(num_iteration):
            rand_sample = self.get_rand_sample(data)
            winner = self.winner(rand_sample)
            self.update(rand_sample,winner,iteration)
            
    def winner(self,x):
        """ Computes the coordinates of the winning neuron for the sample x """
        # get error of x to each weight
        self.activate(x)
        # return index of weight with smallest error; argmin returs the index after flattening the matrix, while unravel_index returs a tuple (row_idx, col_idx) to denote the elem's position
        return unravel_index(self.activation_map.argmin(),self.activation_map.shape)
    
    # x is a single sample, this method returns the error of x to each of the weights(normalized)
    def activate(self,x):
        """ Returns the activation map to x """
        s = x - self.weights
#         self.activation_map = np.sqrt(np.sum(s**2,axis=2)) # no need to bother with normalizing with square root
        self.activation_map = np.sum(s**2,axis=2) 
        return self.activation_map
    
    def update(self,x,win,t):
        """
            Updates the weights of the neurons.
            x - current pattern to learn
            win - position of the winning neuron for x (array or tuple).
            t - iteration index
        """
        # slowly decreasing sigma(regularization) & eta(learning rate)
        eta = self.learning_rate/(1+t/self.T)
        sig = self.sigma/(1+t/self.T)
        g = self.neighborhood(win,sig) * eta
        # every weight in the neighborhood of the winning weight gets updated, with the winning node updating the most
        # the update of each weight is still based on regression, i.e. d/dw(1/2(x-w)^2) = (x-w)
        self.weights += g[:, :, np.newaxis] * (x-self.weights)
#         self.weights /= np.sqrt(np.sum(self.weights**2,axis=2))[:,:,np.newaxis] # no need to bother with normalizing
            
    # returns a multi-dimensional gaussian distribution center in c
    # Note: the normal probability distribution function(PDF) = 1/sqrt(2*pi*sigma^2)*exp(-(x-mean)^2/(2*sigma^2)) 
    # the 1/sqrt(2*pi*sigma^2) ensures the area under the graph=1, since the use of this function is just to create a neighborhood, it can be disregareded
    # to create the whole function across x and y, we need x-mean/y-mean to be an ascending range, with 0 at c[0],c[1] to center it at c
    # which can be achieved by range(x) - c[0], range(y) - c[1]
    # sigma determinese how large the neighborhood is, which is slowly decremented as the model learns
    def neighborhood(self,c,sigma):
        """ Returns a Gaussian centered in c """
        d = 2*pi*sigma*sigma
        ax = exp(-power(self.neigx-c[0],2)/d)
        ay = exp(-power(self.neigy-c[1],2)/d)
        return outer(ax,ay) # the external product gives a matrix

    def distance_map(self):
        """ Returns the average distance map of the weights, i.e. sum of euclidean distance of the surrounding weights to each weight, with max val at 1
            (Each mean is normalized in order to sum up to 1) """
        um = zeros((self.weights.shape[0],self.weights.shape[1]))
        it = nditer(um, flags=['multi_index']) 
        while not it.finished:
            for ii in range(it.multi_index[0]-1,it.multi_index[0]+2):
                for jj in range(it.multi_index[1]-1,it.multi_index[1]+2):
                    if ii >= 0 and ii < self.weights.shape[0] and jj >= 0 and jj < self.weights.shape[1]:
                        um[it.multi_index] += norm(self.weights[ii,jj,:]-self.weights[it.multi_index])
            it.iternext()
        um = um/um.max()
        return um
    
    def win_map(self, data):
        winner_indices = np.zeros((len(data), 2))
        for i, x in enumerate(data):
            winner_indices[i] = self.winner(x)
        return winner_indices

# not used
#     def train_batch(self,data,num_iteration):
#         """ Trains using all the vectors in data sequentially """
#         self._init_T(len(data)*num_iteration)
#         iteration = 0
#         while iteration < num_iteration:
#             idx = iteration % (len(data)-1)
#             self.update(data[idx],self.winner(data[idx]),iteration)
#             iteration += 1

#     def quantization(self,data):
#         """ Assigns a code book (weights vector of the winning neuron) to each sample in data. """
#         q = zeros(data.shape)
#         for i,x in enumerate(data):
#             q[i] = self.weights[self.winner(x)]
#         return q
        
#     def quantization_error(self,data):
#         """ 
#             Returns the quantization error computed as the average distance between
#             each input sample and its best matching unit.            
#         """
#         error = 0
#         for x in data:
#             error += norm(x-self.weights[self.winner(x)])
#         return error/len(data)

#     def win_map(self,data):
#         """
#             Returns a dictionary wm where wm[(i,j)] is a list with all the patterns
#             that have been mapped in the position i,j.
#         """
#         winmap = defaultdict(list)
#         for x in data:
#             winmap[self.winner(x)].append(x)
#         return winmap

#     def activation_response(self,data):
#         """ 
#             Returns a matrix where the element i,j is the number of times
#             that the neuron i,j have been winner.
#         """
#         a = zeros((self.weights.shape[0],self.weights.shape[1]))
#         for x in data:
#             a[self.winner(x)] += 1
#         return a
```

```python
som = Som(16,16,64,sigma=1.0,learning_rate=0.5) # when x/y goes to ~100, training becomes rather slow
som.random_weights_init(data)
print("Initiating SOM.")
som.train_random(data,10000)
print("\n. SOM Processing Complete")

# note: it wouldn't make sense to intialize som with say 100x100 here, since there is only ~2000 examples, if 10000 w are initialized, there is more than enough for each sample to much to an individual weight
```

```python
winners = som.win_map(data)

# the darker parts represent clusters, while the lighter parts represent the division of the clusters
plt.pcolor(som.distance_map().T, cmap = 'bone') 
plt.colorbar()

markers = ['o', 'v', '1', '3', '8', 's', 'p', 'x', 'D', '*']
colors = ["r", "g", "b", "y", "c", (0,0.1,0.8), (1,0.5,0), (1,1,0.3), "m", (0.4,0.6,0)]
for w, label in zip(winners, labels):
    marker, color = markers[label], colors[label]
    plt.plot(
        w[0]+.5, # +0.5 to center the icon in the square
        w[1]+.5,
        marker,
        markerfacecolor='None',
        markeredgecolor=color,
        markersize=12,
        markeredgewidth=2
    )
plt.axis([0,som.weights.shape[0],0,som.weights.shape[1]])
```
